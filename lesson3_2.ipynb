```python
#3*.Напишите код, который будет рисовать график отрезка прямой линии в полярных координатах. 
#Так как прямая имеет заданный угол наклона относительно оси x, 
#то в полярных координатах это будет прямая
def dek_polar(x,y):
    r = pow(pow(x,2)+pow(y,2),0.5)
    phi = np.arctan2(x,y)
    return phi, r


fig = plt.figure(figsize=(5, 5))
x = np.linspace(-2, 6, 1000)
y = list(map(lambda x: 2*x -5, x))
r = []
phi = []
for i in range(len(x)):
    phi_i, r_i = dek_polar(x[i],y[i])
    r.append(r_i)
    phi.append(phi_i)

plt.ylim(-5,5)
plt.xlim(-5,5)
plt.polar(phi, r, color="green")
plt.show()
```

#4.1 Решите систему уравнений:
#y = x^2 – 1
#exp(x) + x∙(1 – y) = 1

```python
from scipy.optimize import fsolve
```

```python
x = np.linspace(-2, 5, 100)
y1 = list(map(lambda x: pow(x,2)-1, x))
y2 = list(map(lambda x: 1 - (1 - np.exp(x))/ x, x))
#plt.ylim(-5,5)
#plt.xlim(-5,5)
plt.plot(x, y1, color="green")
plt.plot(x, y2, color="red")
plt.show()
```

```python
def systema(p):
    x, y = p
    return (y - pow(x,2) + 1, y - (1 - (1 - np.exp(x))/ x))

    
x1,y1 = fsolve(systema, (-2,2))
x2,y2 = fsolve(systema, (2,4))
x3,y3 = fsolve(systema, (4,15))

det = [[x1,y1],[x2,y2],[x3,y3]]
det
```
