### Задание 1
Нарисуйте график функции:
y(x) = k∙cos(x – a) + b
для некоторых (2-3 различных) значений параметров k, a, b

```python
x = np.linspace(-np.pi, np.pi, 201)
k = 1
a = 2
b = 3
plt.plot(x, k*np.sin(x-a) + b, color="green")
plt.plot(x, 2*k*np.sin(x-a-2) + b+3, color="blue")
plt.plot(x, 0.4*k*np.sin(x-a + 2) + b - 1, color="red")
plt.show()
```
