Даны два вектора в трехмерном пространстве: (10,10,10) и (0,0,-10)
Найдите их сумму. (на листочке)
Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
import math
warnings.filterwarnings('ignore')
```

```python
def sum_vek(x,y): # сумма двух векторов
    z = []
    x_rez = x
    y_rez = y
    if len(x)>len(y):
        y_i = []
        for i in range(len(x) - len(y)):
            y_i.append(0)
        y_rez = y + y_i
        x_rez = x
    else:
        x_i = []
        for i in range(len(y) - len(x)):
            x_i.append(0)
        x_rez = x + x_i
        y_rez = y
    for i in range(len(x_rez)):
        z.append(x_rez[i] + y_rez[i])
    return z

def len_vek(x): #длина вектора
    sum_2 = 0
    for i in range(len(x)):
        sum_2 = sum_2 + x[i]*x[i]
    return pow(sum_2, 0.5)
```

Как узнать, принадлежит прямая плоскости или нет?

4.1 Так как плоскость проходит через начало координат, то 0+0+0+D2 = 0, следовательно D2 = 0. Так как для параллельности плоскости необходимо, чтобы все углы сохранялись, то уравнение параллельной плоскости будет:
A1*к*x + B1*к*y + C1*к*z = 0 или A1x + B1y + C1z = 0


4.2 если прямая принадлежит плоскости, то плоскости принадлежат и обе точки, а значит:

A1x1 + B1y1 + C1z1 + D1 = 0 и A1x2 + B1y2 + C1z2 + D1 = 0 или выполняется следующее условие:

A1*(x1-x2) + B1*(y1-y2) + C1*(z1-z2) = 0

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Нарисуйте трехмерный график двух параллельных плоскостей
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5,5,0.5)
Y = np.arange(-5,5,0.5)
X, Y = np.meshgrid(X,Y)
Z = X + Y + 3
Z1 = X + Y + 10
ax.plot_surface(X,Y,Z, edgecolors='g')
ax.plot_wireframe(X,Y,Z1, color='red')
show()
```
