```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
x = np.linspace(-5, 5, 100)
y1 = np.sin(x)
y2 = np.sin(5*x)

fig, ax = plt.subplots(1, figsize=(8, 6))

fig.suptitle('y(k,x) = cos(k*x)', fontsize=15)


ax.plot(x, y1, color="red", label="k=1")
ax.plot(x, y2, color="green", label="k=5")

plt.legend(loc="lower right", title="Legend Title", frameon=False)

plt.show()
```
